# **Activity 11: Distributions and Bayes Algorithm **


## Binomial Distribution

This distribution seems to fluctuate between the values 40 and 60 at random. There is a lot of noise involved and the values don't seem to change based on their order.

```python
import numpy as np

n=100 # number of trials
p=0.5 # probability of success
s=100 # size
binomial = np.random.binomial(n,p,s)
print(binomial)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 100])
plt.plot(binomial,linewidth=2, color='r')
plt.show()
```

## Uniform Distribution

This distribution tends to fluctuate between the values of 1 and 10. Although it appears to do so at random, it seems that there is a reletively similar amount of each value in the distribution.

```python
import numpy as np

uniform = np.random.uniform(low=1, high=10,size=100)
print(uniform)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 10])
plt.plot(uniform,linewidth=2, color='r')
plt.show()
```

## Normal Distribution

This distribution, while noisy, seems to follow a bit of a curve where the center values appear to be more likely to be greater than those on the edges.

```python
import numpy as np
mu, sigma = 0, 0.1 # mean and standard deviation
normal = np.random.normal(mu, sigma, 100)

import matplotlib.pyplot as plt

plt.axis([0, 100, 0, 1])
plt.plot(normal,linewidth=2, color='r')
plt.show()
```

## Bayes Algorithm

Added Multinomial to compare with gaussian.

Ref: https://scikit-learn.org/stable/modules/naive_bayes.html

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB # Gaussian is normal distribution
from sklearn.naive_bayes import MultinomialNB 
X, y = load_iris(return_X_y=True) # https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
gnb = GaussianNB()
mnb = MultinomialNB()
gnb.fit(X_train, y_train)
mnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
y_pred2 = mnb.predict(X_test)
print("Gaussian Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred).sum()))
print("Multinomial Number of mislabeled points out of a total %d points : %d" % (X_test.shape[0], (y_test != y_pred2).sum()))
```
